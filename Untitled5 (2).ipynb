```python
import rasterio
import rasterio.mask
import geopandas as gpd
import matplotlib.pyplot as plt
from rasterio.plot import show
import numpy as np

# Paths
pvout_path = "data/usa_solar/USA_GISdata_LTAy_YearlyMonthlyTotals_GlobalSolarAtlas-v2_GEOTIFF/western-hemisphere/PVOUT.tif"
shapefile_path = "data/usa_shapefile/us-state-boundaries.shp"

# Read USA shapefile (make sure it's in same CRS as raster)
usa = gpd.read_file(shapefile_path)
with rasterio.open(pvout_path) as src:
    if usa.crs != src.crs:
        usa = usa.to_crs(src.crs)

    # Mask raster with USA boundary
    geoms = usa.geometry.unary_union  # combine all states into one geometry
    out_image, out_transform = rasterio.mask.mask(src, [geoms], crop=True)
    out_meta = src.meta.copy()

# Update metadata
out_meta.update({
    "height": out_image.shape[1],
    "width": out_image.shape[2],
    "transform": out_transform
})

# Plot clipped raster
plt.figure(figsize=(12, 8))
cax = plt.imshow(out_image[0], cmap="YlOrRd")
plt.title("Solar PV Potential (PVOUT) - USA", fontsize=14)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
cbar = plt.colorbar(cax, shrink=0.7)
cbar.set_label("kWh/kWp per year")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
cax = ax.imshow(out_image[0], cmap="YlOrRd",
                extent=(out_transform[2], out_transform[2] + out_transform[0]*out_image.shape[2],
                        out_transform[5] + out_transform[4]*out_image.shape[1], out_transform[5]))
usa.boundary.plot(ax=ax, edgecolor="black", linewidth=0.5)  # add state boundaries
plt.title("Solar PV Potential (PVOUT) - USA", fontsize=14)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
cbar = plt.colorbar(cax, shrink=0.7)
cbar.set_label("kWh/kWp per year")
plt.show()
```

```python
!pip install rasterstats
import geopandas as gpd
import rasterio
import rasterstats
import matplotlib.pyplot as plt

states = gpd.read_file("data/usa_shapefile/us-state-boundaries.shp")

raster_path = "data/usa_solar/USA_GISdata_LTAy_YearlyMonthlyTotals_GlobalSolarAtlas-v2_GEOTIFF/western-hemisphere/PVOUT.tif"
raster = rasterio.open(raster_path)

if states.crs != raster.crs:
    states = states.to_crs(raster.crs)

stats = rasterstats.zonal_stats(
    vectors=states,
    raster=raster_path,
    stats="mean",
    nodata=raster.nodata,
    geojson_out=False
)

states["pvout_mean"] = [s["mean"] for s in stats]

fig, ax = plt.subplots(figsize=(20, 14))  # Increase figure size
states.plot(
    column="pvout_mean",
    cmap="YlOrRd",
    linewidth=0.8,
    ax=ax,
    edgecolor="0.8",
    legend=True,
    legend_kwds={"label": "Average PVOUT (kWh/kWp per year)", "shrink": 0.6}
)

ax.set_title("Solar PV Potential (Average PVOUT) by US State", fontsize=20, pad=20)
ax.set_axis_off()

ax.set_xlim(states.total_bounds[[0, 2]])  # minx, maxx
ax.set_ylim(states.total_bounds[[1, 3]])  # miny, maxy

plt.tight_layout()
plt.show()


```
